```python
import pandas as pd
import csv
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

path = r"data\diabetes_012_health_indicators_BRFSS2015.csv"
df = pd.read_csv(path)

print("File:", path)
print("Entries (rows, cols):", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

print("\nSummary statistics (numeric):")
display(df.describe())

print("\nSummary statistics (all columns):")
display(df.describe(include='all'))

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nUnique values per column:")
print(df.nunique())

# Try to locate a likely target column and show its distribution
target_candidates = ['Diabetes_012', 'diabetes_012', 'Diabetes', 'DIABETES', 'diabetes']
target = next((c for c in target_candidates if c in df.columns), None)
if target:
    print(f"\nTarget column '{target}' value counts:")
    print(df[target].value_counts(dropna=False))
    print("\nNormalized distribution:")
    print(df[target].value_counts(normalize=True, dropna=False))

print("\nTop values for object/category columns:")
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(dropna=False).head(10))

print("\nCorrelations (numeric):")
display(df.corr())

# Save a compact column-level summary for later use
summary = pd.DataFrame({
    'dtype': df.dtypes.astype(str),
    'non_null_count': df.count(),
    'missing_count': df.isnull().sum(),
    'unique_count': df.nunique()
})
```

### Compression Function

```python
path = r"data\diabetes_012_health_indicators_BRFSS2015.csv"
df = pd.read_csv(path)
diabetes_parquet = path.replace('.csv', '_parquet.parquet')

for col in df.columns:
    if pd.api.types.is_period_dtype(df[col]):
        df[col] = df[col].astype(str)  # or df[col].dt.to_timestamp() for datetime
    elif df[col].dtype == 'object':
        df[col] = df[col].astype(str)

diabetes_parquet = path.replace('.csv', '_parquet.parquet')
df.to_parquet(diabetes_parquet)



```
